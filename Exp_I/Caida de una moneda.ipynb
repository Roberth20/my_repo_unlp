# Resumen

Segun la teoria cinetica, todos los objetos caen con la aceleracion de la gravedad. De esta forma, en el presente informe se experimenta la validacion experimental del tiempo de caida para una moneda.

# Introduccion

La teoria cinetica proviene directamente de las leyes de Newton, segun las mismas si suponemos que, las dimensiones de la pelota y la velocidad de caída son tal que podemos despreciar el rozamiento con el aire. El movimiento de una partícula con aceleración constante viene dado por (Young et al., 2009, 49)

$$y = h + V_0t - \frac{gt^2}{2}$$

Tomando el caso particular de $h = 0$ y $V_0 = 0$, es decir, altura inicial y velocidad inicial nulas. Podemos despejar el tiempo y encontrar la ecuacion

$$t = \sqrt{\frac{2y}{g}}$$

# Metodo experimental

Para realizar el experimento, se utilizaron:

1. Una cinta metrica (sensibilidad 1 mm)

2. Una tiza

3. Una moneda

4. Un cronometro (sensibilidad de $0.01\ s$)

Para la contruccion de la experiencia, se eligió una pared con un pizarrón como lugar para realizar el experimento. Con la cinta métrica, se midió sobre el costado del pizarrón una altura de 2 metros y con la tiza se marcó dicha altura, para identificar de dónde tirar la moneda una y otra vez (Fig. 1).

```python
from IPython.display import Image

Image(filename="img1.png", width=300)
```

Con el cronómetro en mano, se procedió a dejar caer, un total de 100  tiradas, la moneda desde la altura designada. En cada caso se procedio a colocar la moneda en la posicion adecuada, disminuyendo al minimo las incertidumbres por el pulso de la mano. A su vez, se tomo el tiempo que tardaba en tocar el piso, utilizando el sonido del impacto como referencia.

Los posibles errores involucrados en este proceso provienen del pulso de la mano al momento de soltar la moneda desde el reposo, afectando la altura de la caida en varios milimetros. A su vez, encontramos la incertidumbre conjunta del tiempo de reaccion y velocidad del sonido, implicita de la metodologia manual del experimento. Estas incertezas representan errores accidentales que se minimizaron al maximo posible.

# Resultados

Para examinar la dispersion de las medidas, se construyo un histograma de clases (Fig.2)

```python
# Importamos librerias
import numpy as np
import matplotlib.pyplot as plt

# Cargamos los datos de las medidas
data = np.array([0.85, 0.98, 0.82, 0.88, 0.93, 0.88, 1.27, 0.58, 0.52, 0.58,
                0.68, 1.19, 0.69, 0.48, 1.16, 0.59, 0.59, 0.49, 0.59, 0.89,
                0.69, 0.56, 0.56, 0.77, 0.70, 0.76, 0.56, 0.49, 0.49, 0.63,
                0.56, 0.74, 0.90, 0.70, 0.85, 0.60, 0.63, 0.76, 0.52, 0.73,
                0.69, 0.72, 1.01, 0.61, 0.62, 0.74, 0.72, 0.85, 0.47, 0.52,
                0.53, 0.58, 0.74, 0.60, 0.48, 0.63, 0.79, 0.81, 0.49, 0.63,
                0.49, 0.66, 0.66, 0.67, 0.68, 0.66, 0.75, 0.54, 0.67, 0.48,
                0.63, 0.65, 0.72, 0.41, 0.65, 0.49, 0.82, 0.86, 0.81, 0.54,
                0.57, 0.88, 0.68, 0.63, 0.71, 0.52, 0.66, 0.81, 0.49, 0.54,
                0.71, 0.70, 0.80, 0.70, 0.69, 0.66, 0.77, 0.85, 0.77, 1.05, 0.61], dtype="float64")

data = np.sort(data)
# Obtenemos los datos del histograma y generamos el mismo
y, x, p = plt.hist(data, bins=7)
# Obtenemos valores estadisticos
moda = data[round(len(data)/2)]
media = round(data.mean(), 2)
mediana = round((x[y.argmax()] + x[y.argmax()+1]) / 2, 2)
# Agregamos al histograma los valores
plt.vlines(moda, 0, 35, colors="red", linestyles="dashed", label = f"moda = {moda}")

plt.vlines(media, 0, 35, colors="darkblue", linestyles="-", label = f"promedio = {media}")

plt.vlines(mediana, 0, 35, colors="green", linestyles="-.", label = f"mediana = {mediana}")

# Calculamos la desviacion estandar de la muestra
suma = np.sum((data.mean() - data)**2)
S = np.sqrt(suma/(len(data - 1)))
S = round(S, 2)
plt.annotate(f"Desviacion estandar $S = {S}$", (0.92, 27.5), color = "black")

plt.title("Histograma de 7 clases para el tiempo de caida de una moneda")
plt.xlabel("Tiempo (s)")
plt.ylabel("Conteo")
plt.legend()
plt.show()
```

</p>

Asi, obtenemos el valor medio y la desviación estándar y de esta manera, podemos definir nuestro intervalo como:

$$\bar{x} \pm S = (0.69 \pm 0.16)\ s \qquad \qquad \bar{x} \pm 2S = (0.69 \pm 0.32)\ s$$

A partir de los datos obtenidos, el intervalo $\bar{x} \pm S = (0.69 \pm 0.16)\ s$ contiene 72 medidas, representando el 71.29% de los datos del conjunto. Mientras, $\bar{x} \pm 2S = (0.69 \pm 0.32)\ s$ el intervalo  contiene 97 medidas, las cuales representan el 96.04% de la muestra.

Luego, calculando el numero optimo de medidas  $$N_{opt} = \left( \frac{S}{\sigma_{inst}} \right)^2$$ $$N_{opt} = \left( \frac{0.16}{0.01} \right)^2 = 256$$
Este resultado indica que en la incerteza de la medida, se debe tomar en cuenta la incerteza estadistica $\sigma_{estd}$. De esta manera, la incerteza total, dada por $\sigma = \sqrt{\sigma_{inst}^2 + \sigma_{estd}^2}$ es

$$\sigma = \sqrt{(0.01\ s)^2 + \left( \frac{S}{\sqrt{N - 1}} \right)^2} = \sqrt{(0.01\ s)^2 + \left( \frac{0.16\ s}{\sqrt{99}} \right)^2} = 0.0101\ s$$
$$\sigma \approx 0.01$$

Indicando que la medida es $$t = (0.69 \pm 0.01)\ s$$

# Conclusion

El valor esperado para la medida del tiempo de una moneda qué cae desde 2 metros, dadas las leyes de la cinemática, es de $0.64\ s$.  La medida obtenida, fue indicando que la medida no es exacta, y tiene un error de $0.05\ s$.  Esta variación tiene mayor probabilidad de originarse en la cantidad de medidas tomadas, al ser inferiores al numero optimo, el resultado fue bajo en exactitud.
