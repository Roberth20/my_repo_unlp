# Resumen

Se midió la magnitud de la diferencia longitudinal de un resorte, basado en el
comportamiento del objeto sometido a diferentes masas. Obteniendo así la deformación
sufrida por un objeto sometido a una fuerza, se encontró qué el estiramiento es directamente
proporcional a la fuerza deformante o a la carga. A mayor fuerza, mayor deformación o
desplazamiento.

# Introduccion

La motivación del estudio es analizar las mediciones del desplazamiento de un
resorte, y averiguar cuál es la relación que existe entre su variación longitudinal utilizando
distintas masas. Usando la **ley de Hooke**, el principio físico que determina la elasticidad en
los sólidos, como medio teórico para comparar con los resultados obtenidos.

\begin{equation}
    F = -k\delta
\end{equation}

Donde $k$ es la constante de cada resorte utilizado y $\delta$ la variación de la posición inicial
del resorte (longitud natural) con la longitud final luego de ser excitado con una masa.

Se utilizó las fórmulas para obtener el Chi cuadrado, el Chi cuadrado mínimo y el $R^2_{ajus}$ para poder representar los resultados en una gráfica con una función lineal que corresponda en mayor medida al resultado ($y = a + bx$)

* Chi cuadrado: 
\begin{equation}
    \sum_i^n \left[ f(x_i) - x_i \right]r ^2
\end{equation}

* Chi cuadrado reducido: 
\begin{equation}
    x_v^2 = \frac{\sum_i^n \left[ f(x_i) - x_i \right]r ^2}{N - N_p} \qquad N_p \text{ es el numero de parametros}
\end{equation}

* Coeficiente de ajuste:
\begin{equation}
    R^2_{ajus} = \frac{S^2 - x_v^2}{S^2}
\end{equation}

* Fórmulas de la función lineal con $a$ y $b$ con sus respectivas incertezas ($y = a + bx$)

\begin{align}
    a &= \frac{\sum_{i=1}^N x_i^2 \cdot \sum_{i=1}^N y_i - \sum_{i=1}^N (x_i \cdot y_i) \cdot \sum_{i=1}^N x_i}{N \cdot \sum_{i=1}^N x_i^2 - \left( \sum_{i=1}^N x_i \right)^2} & \Delta a &= \sqrt{\sum_{i=1}^N \left( \frac{\partial a}{\partial y_i} \Delta y \right)^2} = \Delta y \sqrt{\sum_{i=1}^N \left( \frac{\partial a}{\partial y_i} \right)^2}\\
    b &= \frac{N \cdot \sum_{i=1}^N (x_i \cdot y_i) - \sum_{i=1}^N x_i \cdot \sum_{i=1}^N y_i}{N \cdot \sum_{i=1}^N x_i^2 - \left( \sum_{i=1}^N x_i \right)^2} & \Delta b &= \sqrt{\sum_{i=1}^N \left( \frac{\partial b}{\partial y_i} \Delta y \right)^2} = \Delta y \sqrt{\sum_{i=1}^N \left( \frac{\partial b}{\partial y_i} \right)^2}
\end{align}

A cada resorte se lo sometió al mismo procedimiento de estiramiento con una masa de
magnitud desconocida. Una vez ya conocido el comportamiento de $f(x)$ correspondiente a
cada resorte, se puede evaluar el valor de la masa incógnita en los puntos despejando la $x$.
Teniendo en cuenta $f(x) = mx + a$

Con $m$ el estiramiento del resorte por unidad de masa; con $a$ el estiramiento del
resorte sin masa; y $x$ el valor de la masa en distintos puntos.

# Resultados

Los datos obtenidos para cada resorte se graficaron con python. Posteriormente, al observar la distribución de los datos, estos presentaron una relación lineal. Se procedió a desarrollar un modelo que sigue la ecuación lineal, como se puede ver gráficos 1 y 2, el mismo se optimizo reduciendo el $\chi^2$ ( mínimos cuadrados ). 

```python
# Importamos las librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
from scipy.optimize import curve_fit

sea.set()
```

```python
# Cargamos los datos de masas y estiramientos para cada resorte
m1 = np.array([9.9, 20.0, 29.9, 39.9, 49.8, 59.8, 69.9, 79.7, 89.8, 100.0, 109.9])
r1 = np.array([0.1, 0.1, 0.2, 0.3, 0.4, 0.6, 1.0, 1.4, 1.3, 2.3, 2.8])

m2 = np.append(m1.copy(), 120.0)
r2 = np.array([0.3, 0.3, 2.2, 3.6, 4.4, 5.5, 6.8, 8.2, 9.4, 10.7, 12.0, 13.2])

# Creamos la estructura de datos
s1 = pd.DataFrame({"$m_1$": m1, "$\Delta x_1$": r1})
s2 = pd.DataFrame({"$m_2$": m2, "$\Delta x_2$": r2})
```

```python
# Definimos el modelo matematico de la funcion lineal
def lineal(m, a, b):
    return a*m + b
```

```python
# Utilizamos la funcion curve_fit para encontrar los parametros que ajusten el modelo
# a los datos
params, paramsv = curve_fit(lineal, s1["$m_1$"].values[2:], s1.iloc[2:, 1])

# Calculamos los errores de los parametros (Consultar documentacion de curve_fit)
errors = np.sqrt(np.diag(paramsv))

# Graficamos las barras de errores
plt.errorbar(s1.iloc[:, 0], s1.iloc[:, 1], label = "Datos", yerr=0.2, capsize=3, marker="o",
            linewidth=0, elinewidth=1)
# La linea de mejor ajuste
plt.plot(s1["$m_1$"].values[2:], lineal(s1["$m_1$"].values[2:], params[0], params[1]), "purple", label = "Regresion")
# Y la zona de error del modelo
plt.fill_between(s1["$m_1$"].values[2:], lineal(s1["$m_1$"].values[2:], params[0]+errors[0], params[1]+errors[1]),
                lineal(s1["$m_1$"].values[2:], params[0]-errors[0], params[1]-errors[1]), color="purple",
                alpha=0.2)

plt.xlabel("Masas (g)")
plt.ylabel("$\Delta x$ (cm)")
plt.title("Expansion $\Delta x$ del resorte 1 vs masa cargada.")

plt.legend()

plt.show("Resorte 1.jpg")
```

Donde los parametros de ajuste con sus errores son:

```python
print(f"y = ({round(params[0], 3)} +- {round(errors[0],3)})x + ({round(params[1], 1)} +- {round(errors[1], 1)})")
```

```python
# Repetimos procedimiento
params, paramsv = curve_fit(lineal, s2["$m_2$"].values, s2.iloc[:, 1])

errors = np.sqrt(np.diag(paramsv))

plt.errorbar(s2.iloc[:, 0], s2.iloc[:, 1], label = "Datos", yerr=0.2, capsize=3, marker="o",
            linewidth=0, elinewidth=1, ms=3)
plt.plot(s2["$m_2$"].values, lineal(s2["$m_2$"].values, params[0], params[1]), "purple", label = "Regresion",
        alpha=0.5)

plt.fill_between(s2["$m_2$"].values, lineal(s2["$m_2$"].values, params[0]+errors[0], params[1]+errors[1]),
                lineal(s2["$m_2$"].values, params[0]-errors[0], params[1]-errors[1]), color="purple",
                alpha=0.2)

plt.xlabel("Masas (g)")
plt.ylabel("$\Delta x$ (cm)")
plt.title("Expansion $\Delta x$ del resorte 2 vs masa cargada.")

plt.legend()
#plt.show()
plt.show()
```

Y su modelo matematico es:

```python
print(f"y = ({round(params[0], 3)} +- {round(errors[0],3)})x + ({round(params[1], 1)} +- {round(errors[1], 1)})")
```

Para el gráfico del resorte 1, y en orden de obtener una función lineal que ajuste mejor
a los datos, se excluyó del cálculo los primeros 2 valores obtenidos para que nuestro
experimento se acople a un comportamiento lineal. Los mismos tienen la posibilidad de estar
afectados por errores en la ejecución del experimento, ya que debido a la constante del
resorte, las masas pequeñas no influyen en el estiramiento porque no alcanzan el mínimo
punto de quiebre de la constante del resorte, por lo que en los primeros puntos de la gráfica
no se podría evaluar el comportamiento de estiramiento. Una vez obtenidas las funciones
para ambos resortes, se calculó la masa incógnita, además, se formó un gráfico para ilustrar
su ubicación correspondiente que se corresponden con los puntos verdes de las gráficos
consecuentes, 3 y 4.

```python
# Obtenemos los parametros optimizados
params, paramsv = curve_fit(lineal, s1["$m_1$"].values[2:], s1.iloc[2:, 1])
# y sus errores
errors = np.sqrt(np.diag(paramsv))

# Graficamos los errores
plt.errorbar(s1.iloc[:, 0], s1.iloc[:, 1], label = "Datos", yerr=0.2, capsize=3, marker="o",
            linewidth=0, elinewidth=1)

# Definimos la funcion que calculara la masa en funcion del estiramiento
def inverse(x, a, b):
    return (x-b)/a

plt.scatter(inverse([0.1, 0.3, 0.8, 2.8], params[0], params[1]), [0.1, 0.3, 0.8, 2.8], color="green", 
           label="M incognita")

plt.xlabel("Masas (g)")
plt.ylabel("$\Delta x$ (cm)")
plt.title("Expansion $\Delta x$ del resorte 1 vs masa cargada.")

plt.legend()

plt.show()
```

```python
params, paramsv = curve_fit(lineal, s2["$m_2$"].values, s2.iloc[:, 1])

errors = np.sqrt(np.diag(paramsv))

plt.errorbar(s2.iloc[:, 0], s2.iloc[:, 1], label = "Datos", yerr=0.2, capsize=3, marker="o",
            linewidth=0, elinewidth=1, ms=3)

plt.scatter(inverse([2.7, 3.8, 6.3], params[0], params[1]), [2.7, 3.8, 6.3], color="green", 
           label="M incognita", s=10)

plt.xlabel("Masas (g)")
plt.ylabel("$\Delta x$ (cm)")
plt.title("Expansion $\Delta x$ del resorte 2 vs masa cargada.")

plt.legend()

plt.show()
```

Utilizando las medidas de estiramiento de la masa incógnita con cada resorte, y con ayuda de la propagación de incertezas, determinamos que la masa incógnita tiene valores: $m_1 = (36 \pm 26)\ g$ y $m_2 = (34 \pm 7)\ g$. Donde el subíndice denota el resorte utilizado para la medida

# Conclusion
Tras varías observaciones y medidas, se denota que al excitar varios sistemas (resortes) con diferentes cargas de masa, estos expanden sus longitudes. Se encontró que la proporción en la cual se expanden los resortes, sigue una forma lineal, tomando como variable independiente la fuerza ejercida para estirarlo.

Se encontró el valor de la masa incógnita, al introducir las medidas del estiramiento,
influenciados por dicha masa desconocida, dentro de las gráficas de los resortes excitados con
masas de magnitudes conocidas ya medidas y evaluadas. Eligiendo el valor obtenido empleando el resorte 2, debido a que posee menor incerteza.

Al comparar este resultado con la ley de Hooke, vemos que lo encontrado era lo
esperado por dicha ley.
